```python
def ip2long(ip):
    """
    Convert an IP string to long
    """
    packedIP = socket.inet_aton(ip)
    return struct.unpack("!L", packedIP)[0]
```

## Funcion para plotear todos los protocolos de una RTU  
Todas los protocolos que se asocia con la IP de la RTU

```python
def plot_rtu(ip_tst, df,save=False,col=[],intIP=True):
    if intIP: filter ="(srcIP == "+str(ip2long(ip_tst))+" | (dstIP == "+str(ip2long(ip_tst))+"))" 
    else: filter = "(srcIP == '%s' | (dstIP == '%s' )) " %(ip_tst,ip_tst) 
    dfrtu = df[df.eval(filter)]#.reset_index() # df filtered by IP Address
    # obtain protocol names of RTU
    aux_proto = dfrtu[['nDPIMasterProto','nDPIclass']]
    aux_proto = aux_proto.groupby(['nDPIMasterProto','nDPIclass']).value_counts().reset_index()
    protocols = aux_proto['nDPIclass'].to_list() # ['SNMP', 'ICMP', 'TLS', 'DNP3']
    # Lista con nombre de columnas preferidas 
    col_names = list(dfrtu[col].columns.values) # ['timeLast', 'nDPIclass', 'numBytesSnt', 'avePktSize']
    
    df1 = []
    for i in protocols:
        # print("nDPIMasterProto == "+str(i) )    
        evalProto = 'nDPIclass == "%s"' % i
        df1.append( dfrtu[dfrtu.eval(evalProto)][col].reset_index() ) # .reset_index()
    
    subi = np.arange(len(protocols)) 
    x,y = [],[]
    currentColumn = col_names[2]
    for (ix, i,names) in zip(subi,df1,protocols):
        x.append(i['timeLast'])
        y.append(i[currentColumn])
        plt.title(str(ip_tst)+" proto: "+currentColumn)
        plt.plot(x[ix], y[ix], label = "line "+str(names)) 
    plt.legend()    
    plt.show()
```

**Funcion para plotear** un solo protocolo

```python
def plotProtoRtu(ip_tst, df_flows, proto_name, percent=0.8,intIP=True):
    if intIP: filter ="(srcIP == "+str(ip2long(ip_tst))+" | (dstIP == "+str(ip2long(ip_tst))+"))" 
    else: filter = "(srcIP == '%s' | (dstIP == '%s' )) " %(ip_tst,ip_tst)              
    dfrtu = df_flows[df_flows.eval(filter)].reset_index()
    # Obtner Clases que contiene la RTU
    count_proto = dfrtu[['nDPIMasterProto','nDPIclass','numBytesRTAggr']]
    count_proto = count_proto.groupby(['nDPIMasterProto','nDPIclass']).agg({'numBytesRTAggr': 'sum'})
    proto = count_proto.index.get_level_values('nDPIMasterProto')
    protoName = count_proto.index.get_level_values('nDPIclass')
    protoName = protoName.tolist()
    proto = proto.tolist()
    
    if proto_name in protoName: 
        left_e1 = int(len(dfrtu.index)*percent) # 20% visualizacion de los datos
        # Filtrando protocolo por argumento
        protoNumber = proto[protoName.index(proto_name)]
        protFilter = "nDPIMasterProto == " + str(protoNumber)        
        x_plt = pd.to_datetime(dfrtu[dfrtu.eval(protFilter)]['timeLast'].iloc[15:left_e1], unit='s', origin='unix')
        y1 = dfrtu[dfrtu.eval(protFilter)]['numBytesSnt'].iloc[15:left_e1]
        y2 = dfrtu[dfrtu.eval(protFilter)]['numBytesRcvd'].iloc[15:left_e1]
        figrtu, (ax1, ax2) = plt.subplots(2,figsize=(12, 4))
        figrtu.suptitle('Plot bytes send(green) and recived')
        ax1.plot(x_plt, y1, label = "numBytesSnt",color='g' )
        ax2.plot(x_plt, y2, label = "numBytesRcvd")
        plt.legend()
        plt.show()
    else: print("unknow protocol")
    
    
```

```python
plotProtoRtu("9.9.5.237", df300, "DNP3", percent=0.9,intIP=False)
```

```python
def plotcolumns(df1, ip_tst,col, intIP=True ):
    if intIP: filter ="(srcIP == "+str(ip2long(ip_tst))+" | (dstIP == "+str(ip2long(ip_tst))+"))" 
    else: filter = "(srcIP == '%s' | (dstIP == '%s' )) " %(ip_tst,ip_tst) 
    dfrtu = df1[df1.eval(filter)].reset_index()
    # Importar el dataframe
    dataframe = dfrtu[dfrtu.eval("nDPIMasterProto == 244 ")][col]

    # Obtiene las columnas que no son el eje x
    columns = list(dataframe.columns)
    columns.remove("timeLast")
    fheight = len(columns)*2 
         # Crear la figura (width, height)
    plt.figure(figsize=(10, fheight))

    # Crea un subplot para cada columna
    for column in columns:
       plt.subplot(len(columns), 1, columns.index(column) + 1)
       plt.plot(dataframe["timeLast"], dataframe[column])
       plt.title(column)

    plt.savefig('images/ds_columnDNP3proto.png')
    plt.show()
```

## Funcion grafica columnas en misma grafica

```python
def plot_col_proto_rtu(ip_tst, df,col,intIP=True, proto="DNP3",f=1,ff=0):
    if intIP: filter ="(srcIP == "+str(ip2long(ip_tst))+" | (dstIP == "+str(ip2long(ip_tst))+"))" 
    else: filter = "(srcIP == '%s' | (dstIP == '%s' )) " %(ip_tst,ip_tst) 
    filterProto = 'nDPIclass == "%s"' % proto
    dfrtu = df[df.eval(filter)]#.reset_index() # df filtered by IP Address
    dfrtu = dfrtu[dfrtu.eval(filterProto)]#.reset_index# df filtered by PROTOCOL  

    factor = int( len(dfrtu.index)*f)
    factor1 = int( len(dfrtu.index)*ff)
    if dfrtu.empty : print("unknow protocol")      
    else : 
     
        # Lista con nombre de columnas preferidas 
        col_names = list(dfrtu[col].columns.values) # ['timeLast', 'numBytesRcvd', 'numBytesRTAggr', 'maxPktSz', 'avePktSize',
        col_names.remove('timeLast') 
        
        x = dfrtu['timeLast'].iloc[factor1:factor]
        y = []
        
        for columns in col_names:
            y.append(dfrtu[columns].iloc[factor1:factor])
        
        plt.figure(figsize=(12, 4))
        plt.title(str(ip_tst)+" protocol "+ proto)
        
        subi = np.arange(len(col_names)) 
        for iy,name in  zip(y,col_names):
            plt.plot(x, iy, label = "line "+name)
        
        if 'numBytesRTAggr' in col_names: plt.yscale("log")    
        plt.legend()    
        plt.show()
  
        
    
```
